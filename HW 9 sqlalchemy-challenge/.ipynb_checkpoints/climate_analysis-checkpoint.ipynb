```python
from sqlalchemy import create_engine, and_, or_, inspect, func
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
import pymysql
import pandas as pd
import numpy as np
import datetime


dt = datetime.datetime

pymysql.install_as_MySQLdb()
```

```python
Base = automap_base()

engine = create_engine('sqlite:///hawaii.sqlite')
conn = engine.connect()
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)

# Print all of the classes mapped to the base
Base.classes.keys()
```

```python
# Assign the class to a variable
Measurements = Base.classes.measurement
Stations = Base.classes.stations
```

```python
session = Session(engine)
```

# Precipitation Analysis


1. Design a query to retrieve the last 12 months of precipitation data.
2. Select only the date and prcp values.
3. Load the query results into a Pandas DataFrame and set the index to the date column.
4. Plot the results using the DataFrame plot method.
5. Use Pandas to print the summary statistics for the precipitation data.

```python
last_date = session.query(Measurements.date, func.sum(Measurements.prcp)).\
            group_by(Measurements.date).order_by(Measurements.date.desc()).first().date
first_date = (dt.strptime(last_date, "%Y-%m-%d") - datetime.timedelta(365)).date().isoformat()
```

```python
first_date
```

```python
table = session.query(Measurements.date,func.sum(Measurements.prcp)).\
            group_by(Measurements.date).filter(Measurements.date>=first_date).all()
dates, prcps = zip(*table)

prcps = list(prcps)
dates = list(dates)

precipitation = pd.DataFrame(prcps,dates,columns=['precipitation'])
```

```python
import matplotlib.pyplot as plt
import matplotlib.transforms as mtrans
import seaborn as sns

plt.close('all')

sns.set(style='dark')

fig,ax= plt.subplots()

precipitation.plot(linewidth=4,color='#3498db',x_compat=True,ax=ax,grid=True)
plt.yticks(np.arange(0,15,4))
plt.xticks(ha='right')
ax.set_xlabel('date',fontsize=18)
ax.legend(frameon=True,fontsize=15)
ax.grid(color='lightgray')
ax.tick_params(axis='x',rotation=40,labelsize=15,pad=2)
ax.tick_params(axis='y',labelsize=15)

trans = mtrans.Affine2D().translate(20, 0)
for t in ax.get_xticklabels():
    t.set_transform(t.get_transform()+trans)


fig.set_figwidth(6.5)

plt.show()
```

# Station Analysis


1. Design a query to calculate the total number of stations.
2. Design a query to find the most active stations.
3. List the stations and observation counts in descending order
4. Which station has the highest number of observations?
5. Design a query to retrieve the last 12 months of temperature observation data (tobs).
6. Filter by the station with the highest number of observations.

```python
active_station_query =  """
        SELECT m.station, COUNT(m.tobs), s.name
        FROM measurement m
        INNER JOIN stations s
        ON m.station = s.station
        GROUP BY name
        ORDER BY COUNT(m.tobs) DESC
                        """
df_activestations = pd.read_sql(active_station_query, conn)
most_active_station = df_activestations['station'][0]

df_activestations
```

```python
precipitation = pd.DataFrame(prcps,dates,columns=['precipitation'])
precipitation.index.name = 'date'
precipitation = precipitation.reset_index()

dict_prcp = {}
for row in precipitation.iterrows():
    dict_prcp[row[1][0]] = row[1][1]
dict_prcp
```

```python
past_12months_query = """
        SELECT * 
        FROM measurement m
        JOIN stations s
        ON s.station = m.station
        WHERE m.date >= '2016-08-23' AND m.station = '%s'
        """ % most_active_station
df_12month = pd.read_sql(past_12months_query,conn)
df_12month.head()
```

```python
plt.close('all')

fig,ax=plt.subplots()

plt.hist(df_12month['tobs'],bins=12,facecolor='#3498db')
plt.ylabel('Frequency',fontsize=14)
plt.legend(['tobs'],frameon=True,fontsize=12,loc=1)
ax.tick_params(axis='x',labelsize=12,pad=2)
ax.tick_params(axis='y',labelsize=12,pad=2)
plt.grid(color='lightgray')

fig.set_figwidth(5)
plt.show()
```

# Temperature Analysis


1. Write a function called calc_temps that will accept a start date and end date in the format %Y-%m-%d and return the minimum, average, and maximum temperatures for that range of dates.
2. Use the calc_temps function to calculate the min, avg, and max temperatures for your trip using the matching dates from the previous year (i.e. use "2017-01-01" if your trip start date was "2018-01-01")
3. Plot the min, avg, and max temperature from your previous query as a bar chart.
4. Use the average temperature as the bar height.
5. Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr).

```python
def calc_temps(start_date,end_date):
    table = session.query(Measurements.date,Measurements.tobs)\
                .filter(Measurements.date>=start_date,Measurements.date<=end_date).all()
    dates, tobs = zip(*table)

    maximum = max(tobs)
    minimum = min(tobs)
    average = sum(tobs)/len(tobs)
    
    return maximum,minimum,average
start = '2017-02-01'
end = '2017-02-11'   

temp_stats = calc_temps(start,end)

t_range = temp_stats[0]-temp_stats[1]
```

```python
# Plot results here
plt.close('all')

plt.figure().set_figwidth(2)

plt.bar(0.5,temp_stats[2],alpha=0.5,yerr=t_range,facecolor='#FC8D62',width=0.6)

plt.xticks([0,1])
plt.ylim(0,105)
plt.ylabel('Temp (F)')
plt.title('Trip Avg Temp',fontsize=14)
plt.grid(color='lightgray')
plt.gca().axes.get_xaxis().set_visible(False)

plt.show()
```

# Optional Recommended Analysis

1. The following are optional challenge queries. These are highly recommended to attempt, but not required for the homework.
    - Calcualte the rainfall per weather station using the previous year's matching dates.
3. Calculate the daily normals. Normals are the averages for min, avg, and max temperatures.
    - Create a function called daily_normals that will calculate the daily normals for a specific date. This date string will be in the format %m-%d. Be sure to use all historic tobs that match that date string.
    - Create a list of dates for your trip in the format %m-%d. Use the daily_normals function to calculate the normals for each date string and append the results to a list.
    - Load the list of daily normals into a Pandas DataFrame and set the index equal to the date.
    - Use Pandas to plot an area plot (stacked=False) for the daily normals.

```python
def rainfall_by_station(start_date,end_date):
    table = session.query(Measurements.station,Measurements.date,Measurements.tobs)\
                .filter(Measurements.date>=start_date,Measurements.date<=end_date).all()
    
    station_table = session.query(Stations.station,Stations.name)
    
    temp_stations,names = zip(*station_table)
    stations, dates, prcps = zip(*table)

    stations = list(stations)
    prcps = list(prcps)
    dates = list(dates)
    temp_stations = list(temp_stations)
    names = list(names)
    
    
    df = pd.DataFrame({'Station':stations,
                       'Precipitation':prcps,
                       'Date':dates})
    
    df = df.groupby('Station',as_index=False).sum()
    
    df_names = pd.DataFrame({'Station':temp_stations,
                             'Name':names})
    
    df = pd.merge(df,df_names,on='Station',how='inner')
    
    return df
   
start = '2017-02-01'
end = '2017-02-11'

rainfall_by_station(start,end)
```

```python
# date in %m-%d format
# function assumed connection to engine already made

def daily_normals(date):
    date_query = """
            SELECT * 
            FROM measurement m
            JOIN stations s
            ON s.station = m.station
            WHERE m.date LIKE '%-{0}'
            """.format(date)
    
    df_specific_date = pd.read_sql(date_query,conn)
    
    maximum = df_specific_date['tobs'].max()
    minimum = df_specific_date['tobs'].min()
    average = df_specific_date['tobs'].mean()
    
    return [maximum,minimum,average]
    
specific_date = '01-01'
daily_normals(specific_date)
```

```python
def list_of_dates(start_date,end_date):
    table = session.query(Measurements.date)\
                .filter(Measurements.date>=start_date,Measurements.date<=end_date).all()
    dates = pd.Series(list(zip(*table)))
    dates = np.unique(np.array(list(dates)[0]))
    
    return dates.tolist()

listofdates = list_of_dates(start,end)
```

```python
dailyNormals = {'maxes':[],
                'mins':[],
                'avgs':[]}
                
dates_dict = {'dates':[]}

for day in listofdates:
    day = day[-5:]
    stats = daily_normals(day)
    
    dates_dict['dates'].append(day)
    dailyNormals['maxes'].append(stats[0])
    dailyNormals['mins'].append(stats[1])
    dailyNormals['avgs'].append(stats[2])

df_dailies = pd.DataFrame(dailyNormals,index=dates_dict['dates'])
df_dailies
```

```python
plt.close('all')
sns.set(style='dark')
sns.set_context('talk')

fig,ax = plt.subplots()

df_dailies.plot.area(alpha=0.4,linewidth=3,ax=ax,stacked=False)
ax.grid(color='lightgray')
ax.legend(['tavg','tmax','tmin'],frameon=True)
plt.xlabel('date')
ax.tick_params(axis='x',rotation=40,pad=2)
ax.tick_params(axis='y')

trans = mtrans.Affine2D().translate(-15, 0)
for t in ax.get_xticklabels():
    t.set_transform(t.get_transform()+trans)
    
plt.show()
```
